```python
from sympy import * 
```

```python
x = Symbol('x')
```

```python
f = (sin(x)/2 + cos(x)/2)**2 * sin(x)
```

```python
root = sorted(solve(f, x)) # 1.Определить корни
print(root)
```

```python
# Находим экстремумы функции
extremums = sorted(solve(diff(f,x), x))
print(extremums)
```

```python
# 2.Найти интервалы, на которых функция возрастает
# 3.Найти интервалы, на которых функция убывает
# 6.Определить промежутки, на котором f > 0
# 7.Определить промежутки, на котором f < 0

temp = pi
while True:
    if diff(f,x).subs(x, 0) == diff(f,x).subs(x, 0 + temp):
        break
    else:
        temp += temp       
list = extremums[:]
list.append(extremums[0] + temp)
list = sorted(list)
for i in range(1, len(list)):
    if f.subs(x,list[i-1] + 0.000000000000001) < f.subs(x,list[i] - 0.000000000000001):
        print(f'{i}. Функция возрастает на промежутке\n({list[i-1]} : {list[i]})')
    else:
        print(f'{i}. Функция убывает на промежутке\n({list[i-1]} : {list[i]})')
print()
temp = pi
while True:
    if f.subs(x, 1) == f.subs(x, 1 + temp):
        break
    else:
        temp += temp
root.append(root[0] + temp)
for i in range(1, len(sorted(root))):
    if f.subs(x, root[i-1] + 0.000000000000001) > 0:
        print(f'{i}. Функция > 0 на промежутке: ({root[i-1]} : {root[i]})')
    else:
        print(f'{i}. Функция < 0 на промежутке: ({root[i-1]} : {root[i]})')
```
